```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Start Chrome browser
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open the Saramin search page
url = "https://www.saramin.co.kr/zf_user/jobs/list/domestic?cat_kewd=109&loc_mcd=101000&panel_type=&search_optional_item=n&search_done=y&panel_count=y&preview=y"
driver.get(url)

# Wait a few seconds for page to load JS content
time.sleep(5)  # better: use WebDriverWait for production
```

```python
job_elements = driver.find_elements(By.CLASS_NAME, "list_item")

jobs = []
for job in job_elements:
    
    try:
        company_name = job.find_element(By.CLASS_NAME, "company_nm").text
    except:
        company_name = "no company"
    try:
        title = job.find_element(By.CLASS_NAME, "job_tit").text
    except:
        title = "no title"
    try:
        location = job.find_element(By.CLASS_NAME, "work_place").text
    except:
        location = "no location"
    try:
        job_meta = job.find_element(By.CLASS_NAME, "job_meta").text
    except:
        job_meta = "no meta"

    
    
    jobs.append({"Company": company_name, "Title": title, "job_meta": job_meta, "location": location})

# Convert to DataFrame
df = pd.DataFrame(jobs)
df
```

```python
df['Company'][0]
```

```python
datalen = len(df)
for name in range(datalen):
    if df['Company'][name] == 'no company':
        print('row', name, 'deleted')
        df.drop(index=name, inplace=True)
df
```

```python
df.to_csv('ML_engineer.csv', index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Start Chrome browser
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

base_url = "https://www.saramin.co.kr/zf_user/search/recruit?search_area=main&search_done=y&search_optional_item=n&searchType=recently&searchword=ML%20%EC%97%94%EC%A7%80%EB%8B%88%EC%96%B4&recruitPage={page}&recruitSort=relation&recruitPageCount=40&inner_com_type=&company_cd=0%2C1%2C2%2C3%2C4%2C5%2C6%2C7%2C9%2C10&show_applied=&quick_apply=&except_read=&ai_head_hunting="
data = pd.DataFrame()
for page in range(1, 11):  # pages 1 to 10
    url = f"https://www.saramin.co.kr/zf_user/search/recruit?search_area=main&search_done=y&search_optional_item=n&searchType=recently&searchword=ML%20%EC%97%94%EC%A7%80%EB%8B%88%EC%96%B4&recruitPage={page}&recruitSort=relation&recruitPageCount=40&inner_com_type=&company_cd=0%2C1%2C2%2C3%2C4%2C5%2C6%2C7%2C9%2C10&show_applied=&quick_apply=&except_read=&ai_head_hunting="
    print('url', page, 'has been created. getting data...')
    driver.get(url)

    time.sleep(15)

    job_elements = driver.find_elements(By.CLASS_NAME, "item_recruit")

    jobs = []
    for job in job_elements:
        
        try:
            company_name = job.find_element(By.CLASS_NAME, "area_corp").text
        except:
            company_name = None
        try:
            title = job.find_element(By.CLASS_NAME, "job_tit").text
        except:
            title = None
        try:
            location = job.find_element(By.CLASS_NAME, "work_place").text
        except:
            location = None
        try:
            job_condition = job.find_element(By.CLASS_NAME, "job_condition").text
        except:
            job_condition = None
        
        jobs.append({"Company": company_name, "Title": title, "location": location, "job_condition": job_condition})
    
    print('page', page, 'has been scraped')
    df = pd.DataFrame(jobs)
    data = pd.concat([data, df], ignore_index=True)

print('task compelete!')


    
```
