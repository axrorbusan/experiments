```python
#pie
y = np.array([5,10,7,10])
labels=["Apples", "Bananas", "Cherries", "Dates"]
specialdata = [0, 0.1,0,0]
plt.pie(y, labels=labels, startangle=90, explode=specialdata)
```

```python
plt.pie(y, labels=labels, startangle=90, explode=specialdata, shadow=True)
```

```python
plt.pie(y, labels=labels, startangle=90, explode=specialdata, colors=['r','g','b','k'])
```

```python
plt.pie(y, labels=labels, startangle=90, explode=specialdata)
plt.legend()
```
